```
counts = {}


with open('lab01_testFile.txt', 'r') as f:
    for line in f:
      #print(line)
      words = line.split(sep=' ')
      for word in words:
        if word in counts:
          counts[word] += 1
        else:
          counts[word] = 1

counts

```
