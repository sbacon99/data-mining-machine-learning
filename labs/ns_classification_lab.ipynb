```
# Import libraries
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd

import seaborn as sns;
from sklearn.model_selection import train_test_split
from sklearn.neighbors import KNeighborsClassifier
from sklearn.metrics import confusion_matrix
```

```
# Read in dataset and display rows
animals = pd.read_csv('vertebrate2.csv')
animals.head(5)
```

```
# Flying Squirrel Classification

# Probability of being a mammal:
# warm-blooded = 5/5
# gives birth = 5/5
# not aquatic = 4/5
# aerial = 1/5
# has legs = 4/5
# no hibernating = 3/5
# mammal = 5/15
print("Probability of being a mammal:")
mammal = (5/5)*(5/5)*(4/5)*(1/5)*(4/5)*(3/5)*(5/15)
mammal
```

```
# Probability of not being a mammal:
# warm-blooded = 2/10
# gives birth = 1/10
# not aquatic = 3/10
# aerial = 1/10
# has legs = 6/10
# no hibernating = 7/10
# mammal = 10/15
print("Probability of not being a mammal:")
non = (2/10)*(1/10)*(3/10)*(1/10)*(6/10)*(7/10)*(10/15)
non
```

Since P(mammal) > P(not mammal), the new flying squirrel would be classified as a mammal.

```
# Creating ytrain and Xtrain variables 
ytrain = animals["Class"]
Xtrain = animals[["Warm-blooded","Gives Birth","Aquatic Creature","Aerial Creature","Has Legs","Hibernates"]]
```

```
from sklearn.naive_bayes import GaussianNB
clf = GaussianNB()
clf.fit(Xtrain, ytrain)
```

```
# Check accuracy of model
print('Class names:', clf.classes_)
print('Prior probabilities:', clf.class_prior_)
print('Class count:', clf.class_count_)
```

```
testData = [
            ['gila monster',0,0,0,0,1,1,'non-mammals'],
            ['platypus',1,0,0,0,1,1,'mammals'],
            ['owl',1,0,0,1,1,0,'non-mammals'],
            ['dolphin',1,1,1,0,0,0,'mammals']
]
testData = pd.DataFrame(testData, columns = animals.columns)
testData
```

```
# Creating ytest and Xtest variables
ytest = testData["Class"]
Xtest = testData[["Warm-blooded","Gives Birth","Aquatic Creature","Aerial Creature","Has Legs","Hibernates"]]
```

```
# Using model to predict classification of 'Xtest'
ypred = clf.predict(Xtest)
```

```
from sklearn.metrics import classification_report

# Model report
print(classification_report(ytest, ypred))

cm = confusion_matrix(ytest, ypred)
sns.heatmap(cm, square=True, annot=True, fmt='d', cbar=True, cmap="BuGn",
                    xticklabels=clf.classes_,
                    yticklabels=clf.classes_)

plt.xlabel('Predicted Label');
plt.ylabel('True Label')
plt.title('Animal Classification Confusion Matrix')
plt.show()
```
