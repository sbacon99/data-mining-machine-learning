```
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
```

** Dataset was cleaned using textEdit. Duplicate entries as well as entries with negative population values were removed. It did not make sense to alter negative values, so they were excluded entirely. **

```
# Reading in the text file
poptable = np.loadtxt('lab02_populations.csv.txt', delimiter = ',', skiprows=1, dtype = int)
# Ensuring that the table is filled with int's
print(type(poptable[0,0]))
# View imported dataset 
print(poptable)
```

```
# Dimensions of poptable
print('Rows:', poptable.shape[0])
print('Columns:', poptable.shape[1])
```

```
# Descriptive statistics
# Hare
print("Hare descriptive statistics")
print("minimum: ", poptable[:,1].min())
print("maximum: ", poptable[:,1].max())
print("average: ", poptable[:,1].sum() / poptable.shape[0])
print()
# Lynx
print("Lynx descriptive statistics")
print("minimum: ", poptable[:,2].min())
print("maximum: ", poptable[:,2].max())
print("average: ", poptable[:,2].sum() / poptable.shape[0])
print()
# Carrot
print("Carrot descriptive statistics")
print("minimum: ", poptable[:,3].min())
print("maximum: ", poptable[:,3].max())
print("average: ", poptable[:,3].sum() / poptable.shape[0])
```

```
# Hare population > 40,000
mask = poptable[:,1] > 40000
harepop = poptable[mask]
print(harepop[:,0])
```

```
# Plots of population over time
# Curves
year = poptable[:,0]
hare = poptable[:,1]
lynx = poptable[:,2]
carrot = poptable[:,3]

plt.plot(year, hare, marker='o', color='b', label = 'hare')
plt.plot(year, lynx, marker='o', color='r', label = 'lynx')
plt.plot(year, carrot, marker='o', color='g', label = 'carrot')
plt.xticks(np.arange(1900, 1921, 2))
plt.legend( ('hare', 'lynx', 'carrot') )
plt.xlabel('Year')
plt.ylabel('Total Population')
plt.title('Animal Population Over Time')
plt.show()
```

```
# Histogram for carrots
plt.hist(carrot, bins=8)
plt.xlabel('Population')
plt.ylabel('Count')
plt.title('Histogram of Carrot Population')
plt.grid(linewidth = 1)

plt.show()
```

```
# Scatterplot of hares vs lynxes
plt.scatter(hare, lynx, marker='o', color='r', alpha=0.5)
plt.xlabel('Hare Population')
plt.ylabel('Lynx Population')
plt.title('Hares vs Lynxes')
plt.show()
```

Based on the scatterplot, there are no clear signs of linearity. Therefore, there is no evidence of any correlation between hare and lynx populations.

```
# Scatterplot of hares vs lynxes +2
hareMod = poptable[:-2,1]
lynxMod = poptable[2:,2]
carrot = poptable[:,3]

plt.scatter(hareMod, lynxMod, marker='o', color='r', alpha=0.5)
plt.xlabel('Hare Population')
plt.ylabel('Lynx Population (2 years later)')
plt.title('Modified Hares vs Lynxes')
plt.show()
```

```
# set width of bar
barWidth = 0.25

# plot dimensions
plt.figure(figsize=(15, 5))

# Set position of bar on X axis
r1 = np.arange(len(hare))
r2 = [x + barWidth for x in r1]
r3 = [x + barWidth for x in r2]
 
# Make the plot
plt.bar(r1, hare, color='orange', width=barWidth, edgecolor='white')
plt.bar(r2, lynx, color='blue', width=barWidth, edgecolor='white')
plt.bar(r3, carrot, color='green', width=barWidth, edgecolor='white')
 
# Add xticks on the middle of the group bars
plt.xlabel('Year', fontweight='bold')
plt.xticks([r + barWidth for r in range(len(hare))], np.arange(1900, 1921, step=1) )
 

# Create legend & Show graphic
plt.legend( ('hare', 'lynx', 'carrot') )
plt.show()
```
