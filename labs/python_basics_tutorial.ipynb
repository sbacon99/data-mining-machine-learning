## Dictionaries

```
ages = {}
print(ages)

ages["Alice"] = 20
ages["Bob"] = 19
ages["Charlene"] = 21

ages["Alice"] = ages["Alice"] + 1
ages["David"] = 23

if "David" in ages:
  print("Found David")
else:
  print("No David")  
ages
```

```
ages.items()
```

```
for key, value in ages.items():
  print(key, value)
```
